```python
import pandas as pd
df_churn =  pd.read_csv("dataset.csv")
df_churn.head(5)
```

```python
df_churn.columns
df_churn.info()
```

```python
#Retirando espaços antes e depois da str e alterando vazios para null
df_churn["TotalCharges"] = (df_churn["TotalCharges"].str.strip()
                            .replace('', pd.NA))
```

```python
# Convertendo de str para float
df_churn["TotalCharges"].astype(float)
```

```python
# Conferindo quais colunas tem null e alterando de null para 0 na coluna Total Charges
df_churn.isnull().sum()
df_churn["TotalCharges"] = df_churn["TotalCharges"].fillna(0)
```

```python
# Alterei dados de sim e não para número, assim podendo calcular média
convert_calc = ["Partner", "Dependents", "Churn"]

df_churn[convert_calc] = df_churn[convert_calc].replace({"Yes": 1, "No": 0})
df_churn[convert_calc] = df_churn[convert_calc].astype(int)
df_churn.info()
```

```python
# Os 5 cliente que mais pagam mensalmente
df_churn = df_churn.sort_values(by=["MonthlyCharges"], ascending= False)
df_churn.head(5)
```

```python
#Verificação de valores duplicados

df_churn.duplicated().sum()
df_churn.duplicated(subset=["customerID"]).sum()
```

```python
df_churn.head()
```

```python
#Criando a tabela dim_cliente

dim_clientes = df_churn[["customerID","gender","SeniorCitizen","Partner", "Dependents"]] 

dim_clientes.head()
```

```python
#Criando a tabela dim_servicos

dim_servicos = df_churn[["customerID", "PhoneService", "MultipleLines", "InternetService", "OnlineSecurity", "DeviceProtection", "TechSupport", "OnlineBackup", "StreamingTV", "StreamingMovies"]]

dim_servicos.head()
```

```python
#Construindo a tabela dim_contrato

dim_contrato = df_churn[["customerID" ,"Contract", "PaperlessBilling","PaymentMethod"]]

dim_contrato.head()
```

```python
#criando a coluna fato

fato_churn = df_churn[["customerID", "tenure", "MonthlyCharges", "TotalCharges", "Churn"]]

fato_churn.head()
```

```python
#exportando tabelas como csv

dim_clientes.to_csv("dim_clientes.csv", index=False)
dim_servicos.to_csv("dim_servicos.csv", index=False)
dim_contrato.to_csv("dim_contrato.csv", index=False)
fato_churn.to_csv("fato_churn.csv", index=False)
```
